```python
import numpy as np
import matplotlib.pyplot as plt
# Loading the .mat file
from scipy.io import loadmat
data = loadmat("Indian_pines_corrected.mat")               
print(data.keys())                            #cheking what variables are inside the file 
                                              #the actual hyperspectral cube is stored in "indian_pines_corrected"
```

```python
#Extracting the hyperspectral cube
cube = data['indian_pines_corrected']
print(cube.shape)
```

```python
def gamma_correction(image, gamma, c = 1.0):

    # Converting to float
    image = image.astype(np.float32)

    # Normalizing the input pixel values
    R = (image - np.min(image))/(np.max(image) - np.min(image) + 1e-12)

    # Applying gamma correction
    S = c*(R**gamma)

    # Normalizing the output pixel values
    S_norm = (S - np.min(S))/(np.max(S) - np.min(S) + 1e-12)

    return S_norm

```

Applying gamma correction to a single band

```python
band_65 = cube[:, :, 65]  # 65th hyperspectral band
gamma_value = 0.5

gamma_band_65 = gamma_correction(band_65, gamma_value)
```

```python
plt.figure(figsize=(6,6))
plt.title(f"Original Band 65")
plt.imshow(band_65, cmap='gray')
plt.axis('off')

plt.figure(figsize=(6,6))
plt.title(f"Gamma Corrected Band 65 ( = {gamma_value})")
plt.imshow(gamma_band_65, cmap='gray')
plt.axis('off')

plt.show()
```

```python
gamma_value = 3
gamma_band_65 = gamma_correction(band_65, gamma_value)
plt.figure(figsize=(6,6))
plt.title(f"Original Band 65")
plt.imshow(band_65, cmap='gray')
plt.axis('off')

plt.figure(figsize=(6,6))
plt.title(f"Gamma Corrected Band 65 ( = {gamma_value})")
plt.imshow(gamma_band_65, cmap='gray')
plt.axis('off')

plt.show()
```
